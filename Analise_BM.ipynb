# Acesso a energia elétrica da população rural no Brasil

### Filtrando os dados

```python
energia_rur_br1 = dados[
    (dados['Country Code'] == 'BRA') & (dados['Indicator Name'] == 'Access to electricity, rural (% of rural population)')
]
```

```python
energia_rur_br1.info()
```

```python
energia_rur_br = energia_rur_br1.melt(
    id_vars={'Country Name','Country Code','Indicator Name'},
    var_name='year',
    value_name='% of access')

energia_rur_br.sort_values('% of access').head()
```

### Removendo valores ausentes (null)

```python
energia_rur_br.isna().sum()
```

```python
energia_rur_br.notna().sum()
```

```python
energia_rur_br.dropna(inplace=True)
energia_rur_br.notna().sum()
```

### Visualização da % de acesso à eletricidade para a população rural no Brasil (1990 - 2019)

```python
plt.xticks(rotation=45)
sns.lineplot(data=energia_rur_br,
             x='year',y='% of access',
             hue='Country Name',
             style='Country Name',
             markers=True)
```

## Acesso a energia x População rural

### Preparação dos dados

```python
pop_energia_rur = dados[
    (dados['Indicator Name'] == 'Access to electricity, rural (% of rural population)') | 
    (dados['Indicator Name'] == 'Rural population (% of total population)')]
```

```python
pop_energia_rur_melt = pop_energia_rur.melt(
                                    id_vars={'Country Name','Country Code','Indicator Name'},
                                    var_name='year',
                                    value_name='% of access')
```

```python
pop_energia_rur_melt.groupby('Country Code').head()
```

```python
pop_energia_rur_melt.info()
```

```python
pop_energia_rur_melt.astype({'year':int})
```

### Visualização com relplot do Seaborn

```python
sns.set_context('notebook')
sns.relplot(
            x = 'year',
            y = '% of access',
            data = pop_energia_rur_melt[(pop_energia_rur_melt['Country Code'].isin(am_lat)) & (pop_energia_rur_melt.astype({'year':int})['year'] > 1989)],
            hue = 'Indicator Name',
            col = 'Country Name',
            col_wrap = 3,
            kind = 'line')
plt.xticks(rotation = 90)
plt.show()
```

## Relação entre população rural/acesso à eletricidade

### Preparação dos dados

```python
pop_energia_rur_pivot = pd.pivot_table(
    data = pop_energia_rur_melt[pop_energia_rur_melt['Country Code'] == 'BRA'],
    values = '% of access',
    index = ['Country Name','Country Code','year'],
    columns = 'Indicator Name'
    )

df = pop_energia_rur_pivot.dropna()
df.head()
```

```python
x = df.index.get_level_values('year').values

print(x)
```

### Visualização interativa com a biblioteca Bokeh

```python
x = df.index.get_level_values('year').values
y1 = df['Access to electricity, rural (% of rural population)']
y2 = df['Rural population (% of total population)']

output_notebook()
z = figure(title="Acces to electricity X Rural population in Brazil", x_axis_label='year', y_axis_label='%')
z.line(x, y1, color = 'blue',legend_label = 'Access to eletricity',line_width=2)
z.line(x, y2, color = 'red',legend_label = 'Rural population',line_width=2)

show(z)
```

## Criando uma nova coluna no df

```python
pop_energia_rur_pivot['Acces to electricity per Rural population'] = pop_energia_rur_pivot['Access to electricity, rural (% of rural population)']/pop_energia_rur_pivot['Rural population (% of total population)']
pop_energia_rur_pivot.head()
```

```python
pop_energia_rur_pivot.index.get_level_values('year').values
```

```python
y = pop_energia_rur_pivot['Acces to electricity per Rural population']
x = pop_energia_rur_pivot.index.get_level_values('year').values
```

### Visualização da nova coluna com Bokeh

```python
output_notebook()
p = figure(title="Acces to electricity per Rural population", x_axis_label='year', y_axis_label='Acces per Rural Population')
p.line(x, y, line_width=2)
show(p)
```

# Evolução do PIB x Índice de GINI

## No Brasil

### Preparação dos dados

```python
gini_gdp_g = ['Gini index', 'GDP per capita growth (annual %)']
gini_gdp = dados[dados['Indicator Name'].isin(gini_gdp_g)].melt(
                                                                id_vars={'Country Name','Country Code','Indicator Name'},
                                                                var_name='year',
                                                                value_name='values')
gini_gdp.head()
```

```python
gini_gdp_br = gini_gdp[gini_gdp['Country Code'] == 'BRA']
```

```python
y1_gdp = gini_gdp_br[gini_gdp_br['Indicator Name'] == 'GDP per capita growth (annual %)']
```

```python
y2_gini = gini_gdp_br[gini_gdp_br['Indicator Name'] == 'Gini index']
```

### Visualização com Matplotlib

```python
fig, ax = plt.subplots()
ax.plot(
    y1_gdp['year'], 
    y1_gdp['values'],
    color = 'blue')
ax.set_xlabel('year')
ax.set_ylabel('GDP per capita growth', color = 'blue')
ax.tick_params('y',colors = 'blue')

ax2 = ax.twinx()
ax2.scatter(
    y2_gini['year'],
    y2_gini['values'],
    color = 'red')
ax2.set_ylabel('Gini Index', color = 'red')
ax2.tick_params('y',colors = 'red')

ax.set_xticklabels(y1_gdp['year'],rotation = 90)

ax.set_title('GDP x Gini index in Brazil')

plt.show()
```

### Heatmap de correlação

#### Preenchendo os valores ausentes com o método backward fill

```python
pivot_gini_gdp_br = pd.pivot_table(
                                   data = gini_gdp_br,
                                   values = 'values',
                                   index = 'year',
                                   columns = 'Indicator Name',
                                   dropna = True)
```

```python
pivot_gini_gdp_br
```

```python
data_gini_gdp_br_x = pivot_gini_gdp_br.fillna(method = 'bfill')
data_gini_gdp_br_x
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize = (12, 10))
sns.heatmap(data_gini_gdp_br_x.corr())
plt.show()
```

```python
print(data_gini_gdp_br_x.corr())
```

#### Desconsiderando os valores ausentes

```python
data_gini_gdp_br_y = pivot_gini_gdp_br.dropna()
data_gini_gdp_br_y
```

```python
plt.figure(figsize = (12, 10))
sns.heatmap(data_gini_gdp_br_y.corr())
plt.show()
```
